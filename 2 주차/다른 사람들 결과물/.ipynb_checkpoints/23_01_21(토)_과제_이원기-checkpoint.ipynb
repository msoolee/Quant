## Pandas 기본

```python
import pandas as pd
import random
import numpy as np

# DataFrame 생성
df = pd.DataFrame(index=range(0,100), columns=["A","B","C","D"])

# 일괄적으로 value 부여
df = df.apply(lambda _: df.apply(lambda x: str(x.name), axis=1))


# 1. A컬럼의 0번 index에 위치한 값을 선택하시오
df.loc[0] # index name 이용
df.iloc[0]  # index num 이용

# 2. A컬럼의 0~10번 index에 속한 값들을 선택하시오
df.loc[0:10, 'A']

# 3. A컬럼부터 C컬럼까지만 선택하시오
df.loc[:, 'A':'C']


# 랜덤한 데이터 일괄 부여
random.seed(100)
num_df = df.apply(lambda x: df.apply(lambda x: random.randint(0,101),axis=1))

# 4. num_df의 A 와 B값을 합친 D 컬럼을 만드시오.
num_df['D'] = num_df['A'] + num_df['B']

# 5. num_df의 A~D 까지의 sum인 sum 컬럼을 만드시오
num_df["E"] = num_df.sum(axis=1)

# 6. 각 컬럼 값들의 상관계수를 구하시오
def GetCov(col1, col2):
  diff_1 = col1 - col1.mean()
  diff_2 = col2 - col2.mean()
  cov = (diff_1*diff_2).sum() / len(col1)
  return cov

def GetStd(col):
  std = np.sqrt(((col - col.mean())*(col - col.mean())).sum() / len(col))
  return std

raw_df = pd.DataFrame(index=num_df.columns, columns= num_df.columns)
cov_df = raw_df.apply(lambda _: raw_df.apply(lambda x: GetCov(num_df[_.name],num_df[x.name]), axis=1))
corr_df = raw_df.apply(lambda _: raw_df.apply(lambda x: GetCov(num_df[_.name],num_df[x.name]) / (GetStd(num_df[_.name])*GetStd(num_df[x.name])), axis=1))

print(cov_df)
print(corr_df)

# num_df.corr()
```

## Python 람다함수 적용

```python
from time import time
df = pd.DataFrame(index=range(0,100), columns=["A","B","C","D"])

# for문을 이용한 value 부여 (index + col_name)
pos1 = time()
for col in df.columns:
    for ind in df.index:
        df[col].loc[ind] = str(ind) + col.lower()
print(f'For문 소요시간: {time()-pos1}')

# lambda를 이용한 value 부여
pos1 = time()
df = pd.DataFrame(index=range(0,100), columns=["A","B","C","D"])
df = df.apply(lambda _: df.apply(lambda x: str(x.name),axis=1)+ _.name.lower())
print(f'lambda함수 소요시간: {time()-pos1}')

# lambda의 처리속도가 월등히 뛰어남
```

## request + beautifulsoup 이용해서 네이버 주가 가져오기

크롤링 할 페이지: **LG이노텍**  
https://finance.naver.com/item/sise_day.nhn?code=011070&page=1

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt
import requests
from datetime import datetime
# LG이노텍 code=011070

# 시작 날짜에 맞는 페이지 찾기
def FindPage(start_date='20220101', code='011070'):
  start_date = datetime.strptime(start_date, '%Y%m%d')
  finding = True
  page = 0
  agent = 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
  while finding:
    page += 1
    url = f'https://finance.naver.com/item/sise_day.nhn?code=011070&page={page}'
    req = requests.get(url, headers={'user-agent': agent})
    soup = bs(req.text, 'html.parser')
    pages = soup.find_all('span', {'class': 'tah p10 gray03'})
    for i in pages:
      judge = datetime.strptime(i.text, "%Y.%m.%d") <= start_date
      if judge:
        finding = False
  return page

# 가격데이터 크롤링 함수
def GetSise(start_date='20220101', code='011070'):
  empty_df = pd.DataFrame()
  code = code 
  agent = 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
  l_page = FindPage(start_date)

  for page in range(1, int(l_page)+1):
    url = f'https://finance.naver.com/item/sise_day.nhn?code={code}&page={page}'
    req = requests.get(url, headers={'user-agent': agent})
    soup = bs(req.text, 'html.parser')
    table = soup.find('table')
    df = pd.read_html(str(table))[0].dropna()
    empty_df = empty_df.append(df)

  return empty_df

data = GetSise('20220101')
data.sort_values('날짜', inplace=True)
data['Date'] = pd.to_datetime(data['날짜'])
data.index = data['Date'].dt.date
data.drop(['날짜','Date'], axis=1, inplace=True)
```

## matplotlib, 캔들 데이터 그리기

```python
# 종가 데이터 line 그래프
plt.plot(data['종가'])
plt.title('LG Innotech Price')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Price')
```

```python
# Title별 그래프
# For문 이용
Title = ['Close','Percent Change','Open','High','Low']
for title,i in enumerate(data.columns[:-1]):
    plt.plot(data[i])
    plt.title(Title[title])
    plt.xlabel('Date')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# plt.subplots 이용
rows = 3
cols = 2
fig, axs = plt.subplots(rows, cols, figsize=((18, 15)))
for i, title in enumerate(data.columns[:-1]):
  axs[int(i/cols), i%cols].plot(data[title])
  axs[int(i/cols), i%cols].set_title(Title[i])
  axs[int(i/cols), i%cols].tick_params(axis='x', rotation=45)
```
