```python
# 패키지 임포트
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## 1. 포트폴리오 구성하기

**핵심-위성 전략으로**, 핵심 종목으로 안정적인 수익을, 위성 종목으로 초과 수익을 창출해 수익성, 안정성을 챙기는 전략  
또한 **주식 60, 채권 40 전략**도 동시에 채택해 봄  
즉, 주식 60 중 핵심 90 위성 10(신사업,성장성이 높은 사업 등), 채권 40 중 핵심 90 위성 10 (회사채, 장기국고채 등) 으로 구성하려 했으나  
먼저 종목별로 동일비중 포트폴리오를 구성해봄

주식 핵심 - 지수 ETF[SPY], 로우볼[SPLV], 리튬배터리[LIT], 배당[SCHD], 금[SGOL] / 위성 - 밀레니얼[MILN], IOT[SNSR]   
채권 핵심 - 단기채[LQD] / 위성 - 장기채[TLT]  
밀레니얼 -> 밀레니얼 세대의 소비와 관련된 종목들로 이루어진 ETF  
원유같은 경우 최근 특수상황으로 인한 변동성이 너무 심해 종목에서 제외  

```python
tickers = ['SPY', 'SPLV', 'LIT', 'SCHD', 'SGOL', 'MILN', 'SNSR', 'SHY', 'LQD', 'TLT']
etf = yf.Tickers(tickers)
data = etf.history(start='2016-01-01',actions= False)['Close']
data = data.resample('3M').last()
```

```python
# 데이터 시각화
# 2016년 7월 31일까지 MILN과 SNSR에서 Nan값이 발생, 이후의 데이터로만 그래프로 그려본다
l_data = data.iloc[3:, :]
P0 =  l_data.iloc[0]

(l_data / P0 * 100).plot(figsize = (15, 6))
plt.show()
```

## 2. 동일비중 포트폴리오 수익률 계산하기

```python
# 포트폴리오 수익률 계산

rets = l_data.pct_change().fillna(0)
parm_rets = rets.mean(axis=1)
parm_cump_rets = (1 + parm_rets).cumprod() - 1
print(f'2016-10-31 ~ 2023-01-31 누적 수익률: {parm_cump_rets[-1]}')
```

```python
# 포트폴리오 수익률 시각화
plt.plot(parm_cump_rets)
plt.title('EW Backtest')
plt.xlabel('Date')
plt.ylabel('Returns')
```

## 3. 개선된 포트폴리오 수익률 계산하기

```python
# 주식 6 채권 4의 비율을 유지하되 core / settle 9:1 전략을 혼합해 사용

def make_weight_list(tickers, port_dict, weight_dict):
  weight_list = []

  for ticker in tickers:
    for key, val in port_dict.items():
      if ticker in val:
        weight = weight_dict[key] / len(val)
        weight_list.append(weight)
    
  return weight_list

port_dict = {
    'core_stock': ['SPY', 'SPLV', 'LIT', 'SCHD', 'SGOL'],
    'settle_stock': ['MILN', 'SNSR'],
    'core_bond': ['SHY'],
    'settle_bond': ['LQD', 'TLT'],
}

weight_dict = {
    'core_stock': 0.6*0.9,
    'settle_stock': 0.6*0.1,
    'core_bond': 0.4*0.9,
    'settle_bond': 0.4*0.1,
}

# 주식핵심, 주식위성, 채권핵심, 채권위성이므로 가중치를 부여한 뒤 포트폴리오 수익률 계산
weight = make_weight_list(tickers, port_dict, weight_dict)
weight_rets = rets.dot(weight)

# 포트폴리오 누적 수익률
dw_cump_rets = ((weight_rets + 1).cumprod() - 1)
print(f'2016-10-31 ~ 2023-01-31 누적 수익률: {dw_cump_rets[-1]}')

plt.plot(dw_cump_rets)
plt.title('DW Backtest')
plt.xlabel('Date')
plt.ylabel('Returns')
```

동일비중보다 개선된 수익률을 얻을 수 있었음.
