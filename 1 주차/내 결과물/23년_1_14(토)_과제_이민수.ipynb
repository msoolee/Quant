## 3. 동일 비중 수익률 계산

```python
#색깔 팔레트
pal = sns.color_palette('Spectral',len(tickers))
```

```python
#가중치 계산 함수                ######################
def get_ew_weights(rets):
  noa = len(rets.columns)
  weights = np.ones_like(rets) * (1/noa)
  return weights

weights = get_ew_weights(rets)[-1]
weights                          ######################
```

```python
# 파이차트 시각화
plt.figure(figsize=(10, 10))
wedgeprops = {'width': 0.32, 'edgecolor': 'w', 'linewidth': 3}
plt.pie(weights, labels=tickers, autopct='%.2f%%', wedgeprops=wedgeprops, colors=pal)
plt.show()
```

```python
port_rets = rets.mean(axis =1)

# 포트폴리오 누적 수익률
port_cum_rets = (1 + port_rets).cumprod() - 1

# 그래프 결과
plt.plot(port_cum_rets)
plt.title('EW Backtest')
plt.xlabel('Date')
plt.ylabel('Returns')

#MDD
'''
MDD 계산! => 최대 수익률 점과 최소 수익률 점을 찾아서 return값으로 반환시킨다
이때 주의할 점은 최대 수익률 점이 최소 수익률 점보다 왼쪽 편에 있어야된다!
'''
def get_mdd(x):
    """
    MDD(Maximum Draw-Down)
    :return: (peak_upper, peak_lower)
    """
    arr_v = np.array(x)
    peak_lower = np.argmax(np.maximum.accumulate(arr_v) - arr_v)
    peak_upper = np.argmax(arr_v[:peak_lower])
    return peak_upper, peak_lower

plt.scatter(port_cum_rets.index[get_mdd(port_cum_rets)[0]],port_cum_rets[5], s = 100, c = 'r')
plt.scatter(port_cum_rets.index[get_mdd(port_cum_rets)[1]],port_cum_rets[11], s = 100, c = 'b')



print(f'2016-10-31 ~ 2023-01-31 누적 수익률: {port_cum_rets[-1]*100} % ')
print( '최대 수익률 낙폭: ' + '\033[95m'+ str(round(port_cum_rets[5]* 100,4)) +'%'+ ' -> '+ '\033[96m'+ str(round(port_cum_rets[11]* 100,4)) + '%')
```

## 4. 비중 조정 후 수익률 계산

##### 비중 조정 전략

 4개의 성장투자성 자산에 50% 비율을 부여함으로써 <br>
 조금더 공격적인 전략으로 바꾼다. <br>
 원래 안정 : 공격 = 6:4 였던것을 5:5로 바꾼다.

```python
# 4개의 성장투자성 자산에 50% 비율을 부여함으로써 
# 조금더 공격적인 전략으로 바꾼다.

#가중치 계산 함수                ######################
def make_weight_list(tickers, port_dict, weight_dict):
  weight_list = []

  for ticker in tickers:
    for key, val in port_dict.items():
      if ticker in val:
        weight = weight_dict[key] / len(val)
        weight_list.append(weight)
    
  return weight_list

port_dict = {
    'stables' : ['SPY','GLD','BIL','^FVX','^TNX','114820.KS'],
    'agressives' : ['QQQ','XLK','DRIV','IHI']
}

weight_dict = {
    'stables': 0.5,
    'agressives': 0.5
} 



weight = make_weight_list(tickers, port_dict, weight_dict)

weight_rets = rets.dot(weight)

# 파이차트 
plt.figure(figsize=(10, 10))
wedgeprops = {'width': 0.32, 'edgecolor': 'w', 'linewidth': 3}
plt.pie(weight, labels=tickers, autopct='%.2f%%', wedgeprops=wedgeprops, colors=pal)
plt.show()
```

```python



dw_cump_rets = ((weight_rets + 1).cumprod() - 1)

#MDD 계산
plt.scatter(dw_cump_rets.index[get_mdd(dw_cump_rets)[0]],dw_cump_rets[5], s = 100, c = 'r')
plt.scatter(dw_cump_rets.index[get_mdd(dw_cump_rets)[1]],dw_cump_rets[11], s = 100, c = 'b')


print(f'2016-10-31 ~ 2023-01-31 누적 수익률: {dw_cump_rets[-1]*100} % ')
print( '최대 수익률 낙폭: ' + '\033[95m'+ str(round(dw_cump_rets[5]* 100,4)) +'%'+ ' -> '+ '\033[96m'+ str(round(dw_cump_rets[11]* 100,4)) + '%')

plt.plot(dw_cump_rets)
plt.title('DW Backtest')
plt.xlabel('Date')
plt.ylabel('Returns')

```
