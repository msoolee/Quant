##1-1 Moving Average (SMA, EMA) 정의

이동평균선(이하 이평)은 과거의 가격이 미래의 가격추세에 영향을 줄 수 있다는 것과 (통계학적으로 AUTO CORRELATION이라고 함)시장엔 추세가 있음을 기본적인 전제로 삼고 분석하는 투자기법인 것이다. 이평은 평활기법(SMOOTHING TECHNICS)중에서 가장 널리 알려진 기법으로 추세의 전환 시점을 신속하게 파악하기 위해 사용되며 매일의 가격이 이평의 진행 방향과 어떤 상관관계를 나타내는가를 분석하여 향후의 가격 동향을 예측 하고자 하는 지표이다. 즉, 가격 이평선에 의한 주가 예측의 방법으로는 가격 이동평균선 자체의 방향성과 이평선과 현실 주가의 괴리도가 이론의 중심이 되고 있다.

이평은 추세 전환 국면인 **보합에서 하락 추세로** 전환 되는 경우나 **보합에서 상승 추세로** 전환하는 경우, **상승에서 보합으로**, **하락에서 보합으로** 전환하는 경우에 아주 유효한 지표며 ***보합 추세에서는 매우 부적절하다.*** 

보합추세에서 나쁜 이유는 만일 시장가격이 갑자기 급락하거나 급등할 경우, 이런 과정이 되풀이 된다면 시장의 급등을 부분적으로 반영하여 이평선이 가격 상승 추세 신호를 보일때 가격은 급락하게 되고 반대로 가격이 급락하여 이평선이 가격하락 추세를 보이면 가격이 급등하여 이평선은 시장을 제대로 반영하지 못할 것이다. 즉, 이평은 장기적으로 시장이 뚜렷하게 상승, 하락추세를 보이는 경우에 적절하게 사용 되어야 하며 다른 추세 전환 국면에서도 이러한 이유 때문에 가격 예측이 매우 용이하다.

이동 평균선은 주가의 변동폭을 좀 더 유연하게 한 것이므로, 주가는 이동 평균선과 균형을 이루면서 변동하는 것이 보통이다. 따라서 이동 평균선을 활용하여 주가를 분석하고자 할 때는 다음과 같은 성질을 파악하는 것이 중요하다.
- 강세 시장에서는 주가가 이동 평균선 위에서 파동을 거듭하면서 상승한다.
- 약세 시장에서는 주가가 이동 평균선 밑에서 파동 운동을 하면서 하락한다.
- 보합 국면의 애매한 주가는 이동 평균선과 밀착하여 파동 운동을 하다가 이동 평균선을 이탈한다.
- 상승하고 있는 이동 평균선을 주가가 하향 돌파할 때는 조만간 반전 하락할 가능성이 크다.
- 하락하고 있는 이동 평균선을 주가가 상향 돌파할 때에는 곧 반전할 가능성이 크다.
- 이동 평균하는 기간이 길면 길수록 이동 평균선은 더욱 유연해진다.
- 주가가 이동 평균선으로부터 지나치게 떨어져 있을 때는 이동 평균선으로 되돌아오는 경향이 있다.
- 주가가 이동 평균선을 밑에서 위로 상향 돌파할 때는 매입 신호이며, 위에서 밑으로 하향 돌파할 때는 매도 신호가 된다.

위에서와 같은 기본 성질을 파악한 후에 매입·매도 신호를 찾는 것이 중요한 데, 매매 신호는 시장의 상황(강세 시장, 약세 시장)에 따라 다르므로 각 국면을 참고해서 찾아야 한다. 예컨대, 강세 국면의 초기인지 중간인지 마지막 국면인지, 보합 국면인지, 약세 국면의 초기인지 중간인지 최종 국면인지를 파악해야 하나, 이평선 만으로는 파악할 수 없다. 

따라서 하나의 지표만을 이용해 매매를 하면 위험하다. 다른 지표들을 참고해서 실전에 적용해야된다.


장·단기 이동 평균선은 각각 장·단점을 가지고 있다. 중·장기 이동 평균선은 매일 매일의 주가 변동에 거의 좌우되지 않기 때문에, 주가의 대세를 판단하는 데에는 유용하나 시세를 신속히 판단하는 데는 어려운 점이 있다. 단기 이동 평균선은 시세의 전환을 빨리 나타내 주는 반면, 주가의 미미한 변동에도 민감하게 반응하므로 매매 시점을 포착하기가 어렵다는 결점이 있다. 그러므로 장기선, 중기선 및 단기선의 성질과 특징을 각각 잘 파악하여야 매매 타이밍을 정확히 포착할 수 있다

```python
#Simple Moving Average 계산

import FinanceDataReader as fdr
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline

ticker = ['011930']
df = fdr.DataReader('011930', '2018-01-01')

#SMA를 구하고, 종가와 합친 테이블을 만든다. 

# SMA 5days
sma5 = df['Close'].rolling(window=5).mean() #5일동안의 종가를 평균을 낸 것이 rolling(5).mean()
df_sma5 = sma5.to_frame(name='SMA5') # Series to DataFrame
# SMA 20days
sma20 = df['Close'].rolling(window=20).mean()
df_sma20 = sma20.to_frame(name='SMA20') # Series to DataFrame
#SMA 60days
sma60 = df['Close'].rolling(window=60).mean()
df_sma60 = sma60.to_frame(name='SMA60') #Series to DataFrame
#SMA 120days
sma120 = df['Close'].rolling(window=120).mean()
df_sma120 = sma120.to_frame(name='SMA120') #Series to DataFrame


# Concatenate SMA5 and SMA20
df_sma = pd.concat([df['Close'], df_sma5, df_sma20,df_sma60,df_sma120], axis=1)
# Remove NaN
df_sma = df_sma.dropna()
```

```python
#차트를 그려보자.
# Draw Chart for SMA
s_date = '2022-01-01'
e_date = '2023-02-05'

fig, ax = plt.subplots(figsize=(20,12))
ax.plot(df_sma.loc[s_date:e_date, :].index, df_sma.loc[s_date:e_date, 'Close'], label='Price',color = "blue", linewidth = "3.0")
ax.plot(df_sma.loc[s_date:e_date, :].index, df_sma.loc[s_date:e_date, 'SMA5'], label='5 SMA',color='green')
ax.plot(df_sma.loc[s_date:e_date, :].index, df_sma.loc[s_date:e_date, 'SMA20'], label='20 SMA',color='orange')
ax.plot(df_sma.loc[s_date:e_date, :].index, df_sma.loc[s_date:e_date, 'SMA60'], label='60 SMA',color='red')
ax.plot(df_sma.loc[s_date:e_date, :].index, df_sma.loc[s_date:e_date, 'SMA120'], label='120 SMA',color='brown')
ax.legend(loc='best')
ax.set_title('Simple Moving Average')
ax.set_ylabel('Price')
```

```python
#EMA 계산 (EMA 계산할때 df.ewm() 함수 이용한다!)

# EMA 5days
ema5 = df['Close'].ewm(span=5,min_periods=5).mean() #5일동안의 종가를 평균을 낸 것이 rolling(5).mean()
df_ema5 = ema5.to_frame(name='EMA5') # Series to DataFrame
# EMA 20days
ema20 = df['Close'].ewm(span=20,min_periods=20).mean()
df_ema20 = ema20.to_frame(name='EMA20') # Series to DataFrame
#EMA 60days
ema60 = df['Close'].ewm(span=60,min_periods=60).mean()
df_ema60 = ema60.to_frame(name='EMA60') #Series to DataFrame
#EMA 120days
ema120 = df['Close'].ewm(span=120,min_periods=120).mean()
df_ema120 = ema120.to_frame(name='EMA120') #Series to DataFrame


# Concatenate EMA5 and EMA20
df_ema = pd.concat([df['Close'], df_ema5, df_ema20,df_ema60,df_ema120], axis=1)
# Remove NaN
df_ema = df_ema.dropna()

fig, ax = plt.subplots(figsize=(20,12))
ax.plot(df_ema.loc[s_date:e_date, :].index, df_ema.loc[s_date:e_date, 'Close'], label='Price',color = "blue", linewidth = "3.0")
ax.plot(df_ema.loc[s_date:e_date, :].index, df_ema.loc[s_date:e_date, 'EMA5'], label='5 EMA',color='green')
ax.plot(df_ema.loc[s_date:e_date, :].index, df_ema.loc[s_date:e_date, 'EMA20'], label='20 EMA',color='orange')
ax.plot(df_ema.loc[s_date:e_date, :].index, df_ema.loc[s_date:e_date, 'EMA60'], label='60 EMA',color='red')
ax.plot(df_ema.loc[s_date:e_date, :].index, df_ema.loc[s_date:e_date, 'EMA120'], label='120 EMA',color='brown')
ax.legend(loc='best')
ax.set_title('Exponential Moving Average')
ax.set_ylabel('Price')
```

###1-2 Moving Average Convergence Divergence

MACD의 원리는 단기이평과 장기이평선의 차이(DIVERGENCE)가 커지게(DIVER) 되면 언제인가는 반드시 차이가 원래대로 수렴(CONVERGE) 되는 경향이 있으므로 이러한 가격 변동의 성질을 이용하여 매수, 매도 기회를 포착하려는 시도에서 개발된 기법이다.

**MACD = 단기이평12 - 장기이평26**

**MACD SIGNAL LINE = Difference (단기이평12 - 장기이평26) * 9**  

MACD는 이동평균의 문제점인 시차(TIME LAG)의 문제를 어느 정도 해결 가능한데 예를 들어 10일 이평선이 존재한다면 일일 가격 변동량은 이평선에 1/10의 영향밖에 미치지 못할 것이다. 그러나 MACD는 단기, 장기 이평선의 차이가 가장 큰 시기에 거래신호를 나타내기 때문에 어느 정도는 해결 가능 하다고 볼 수 있다. 바로 이 차이의 움직임을 파악하기 위해 시그널 라인이라는 지표를 같이 사용하게 되는 것이다. 즉, 12일간과 26일간의 지수평균을 구한 후 이 두지수 평균의 차이를 다시 9 일간의 지수 평균으로 산출하는 것이다.

상승추세(UP TREND)에서는 단기 이평선이 장기 이평선의 위에 있으므로 MACD는 양의 값을 나타내며, 하락 추세에서는(DOWN TREND)에서는 단기 이평선이 장기 이평선의 아래에 있게 되므로 MACD가 음의값을 나타내게 된다. MACD가 음에서 양, 양에서 음으로 전환되는 시점이 단기와 장기 이평선이 교차되는 시점이므로 MACD가 양의 값을 나타내는 경우에는 상승추세라 할 수 있고 음의 값을 나타낸다면 하강 추세라 할 수 있다.

```python
import plotly.graph_objects as go #MACD 그래프 그리기 위해 필요한 라이브러리

ticker = ['011930','086520','005930']
df = {}
for t in ticker:
  data = fdr.DataReader(t, '2022-01-01')
  data.dropna(how='any',inplace=True)
  df[t] = data

def MACD(DF, a=12 ,b=26, c=9):
    """function to calculate MACD
       typical values a(fast moving average) = 12; 
                      b(slow moving average) =26; 
                      c(signal line ma window) =9"""
    df = DF.copy()
    df["ma_fast"] = df["Close"].ewm(span=a, min_periods=a).mean()
    df["ma_slow"] = df["Close"].ewm(span=b, min_periods=b).mean()
    df["macd"] = df["ma_fast"] - df["ma_slow"]
    df["signal"] = df["macd"].ewm(span=c, min_periods=c).mean()
    df["macdDiff"] = df["macd"] - df["signal"]
    return df.loc[:,["macd","signal",'macdDiff']]

for t in ticker:
    df[t][["MACD","SIGNAL",'macdDiff']] = MACD(df[t], a=12 ,b=26, c=9)

```

##1-3 ADX 지수

Average Directioinal Movement Index 는 가격이 횡보 국면에 있을 때는 0에 가까운 값을 가지게 되고 상승 혹은 하락추세가 강하게 보일수록 ADX의 값은 커지게 된다.
이것은 하락과 상승을 반복하는 단순한 곡선이지만 추세의 시작과 그 끝, 그리고 추세가 계속 진행 중에 있는지를 알려주는 유용한 지표이다. 또한 하락과 상승중인 추세의 강도를 나타내므로 PDI, MDI와 병행 사용하면 유용한 매매 판단 지표로 사용할 수 있다. ADX가 큰 값을 보이면 현재의 **추세 강도가 강하게** 나타내는 것을 의미하며 ADX의 값이 적으면 적을수록 현재의 **추세 강도가 약하다**는 것을 의미한다.
이 지표는 상승, 혹은 하락하는 추세의 강도만을 나타내므로 **PDI 와 MDI 를 반드시 같이 사용**하여야 하고 ADX 의 추세를 분석하기 위하여 ADX 의 10일 단순 이동평균 값을 구하여 골든, 데드 크로스의 방법을 사용하는 매매 방법이 있다.
ADX 에서 Directional의 의미는 상승이나 하락 모두를 의미하므로 ADX가 증가한다고 반드시 가격이 상승추세를 의미하는 것은 아니다. ADX는 단지 추세의 강도만을 나타낼 뿐이며 ADX지표 자체로는 상승추세인지 하락 추세인지를 알 수가 없는 것이다.

추세의 강도를 측정하기 위한 지표이므로 단일지표로 매매시기를 포착하는 것보다는 추세의 확인을 하는 보조지표로 사용하는 것이 좋다.
ADX값이 저점에서 상승을 시작했는데 주가가 하락하는 역전이 시작되는 경우에는 주가가 상당폭 하락할 것을 예상하여 추가적인 매도에 가담해야 한다.
ADX값이 고점을 형성하고 하락하면서 주가가 상승하는 역전 현상이 타날 경우에는 하락하던 주가가 반등을 시도하는 것으로 보고 단기매매에 국한해야 한다.
3번처럼 하락하던 ADX 값이 재상승하면서 주가가 계속적으로 상승하는 경우는 자주 나타나지는 않지만 상승전환한다고 판단하면 된다.

```python
import yfinance as yf
# Download historical data for required stocks
tickers = ["AMZN","GOOG","MSFT"]
ohlcv_data = {}

# looping over tickers and storing OHLCV dataframe in dictionary
for ticker in tickers:
    temp = yf.download(ticker, start='2022-01-01')
    temp.dropna(how="any",inplace=True)
    ohlcv_data[ticker] = temp
    
def ATR(DF, n=14):
    "function to calculate True Range and Average True Range"
    df = DF.copy()
    df["H-L"] = df["High"] - df["Low"]
    df["H-PC"] = abs(df["High"] - df["Adj Close"].shift(1))
    df["L-PC"] = abs(df["Low"] - df["Adj Close"].shift(1))
    df["TR"] = df[["H-L","H-PC","L-PC"]].max(axis=1, skipna=False)
    df["ATR"] = df["TR"].ewm(com=n, min_periods=n).mean()
    return df["ATR"]

def ADX(DF, n=14):
    "function to calculate ADX"
    df = DF.copy()
    df["ATR"] = ATR(DF, n)
    df["upmove"] = df["High"] - df["High"].shift(1)
    df["downmove"] = df["Low"].shift(1) - df["Low"]
    df["+dm"] = np.where((df["upmove"]>df["downmove"]) & (df["upmove"] >0), df["upmove"], 0)
    df["-dm"] = np.where((df["downmove"]>df["upmove"]) & (df["downmove"] >0), df["downmove"], 0)
    df["+di"] = 100 * (df["+dm"]/df["ATR"]).ewm(alpha=1/n, min_periods=n).mean()
    df["-di"] = 100 * (df["-dm"]/df["ATR"]).ewm(alpha=1/n, min_periods=n).mean()
    df["ADX"] = 100* abs((df["+di"] - df["-di"])/(df["+di"] + df["-di"])).ewm(alpha=1/n, min_periods=n).mean()
    return df["ADX"]

for ticker in ohlcv_data:
    ohlcv_data[ticker]["ADX"] = ADX(ohlcv_data[ticker],20)
```

##1-4 Schaff Trend: A Faster and more Accurate Indicator


The Schaff Trend Cycle (STC) is a charting indicator that is commonly used to identify market trends and provide buy and sell signals to traders. Developed in 1999 by noted currency trader Doug Schaff, STC is a type of oscillator and is based on the assumption that, regardless of time frame, currency trends accelerate and decelerate in cyclical patterns.

Many traders are familiar with the moving average convergence/divergence (MACD) charting tool, which is an indicator that is used to forecast price action and is notorious for lagging due to its slow responsive signal line . By contrast, STC’s signal line enables it to detect trends sooner. In fact, it typically identifies up and downtrends long before MACD indicator.

**While STC is computed using the same exponential moving averages as MACD, it adds a novel cycle component to improve accuracy and reliability. While MACD is simply computed using a series of moving average, the cycle aspect of STC is based on time (e.g., number of days).**

While the STC indicator seems to boast higher reliability than MACD, it has some inherent flaws. Namely, it can linger in overbought and oversold territory for extended periods of time. For this reason, the indicator is most often used for its intended purpose of following the signal line up and down, and taking profits when the signal line hits the top or bottom. Let's see how it works.

While MACD generates its signal when the MACD line crosses with the signal line, the STC indicator generates its buy signal when the signal line turns up from 25 (to indicate a bullish reversal is happening and signaling that it is time to go long), or turns down from 75 (to indicate a downside reversal is unfolding and so it's time for a short sale).

##2-1 RSI (relative strength index)

RSI는 위의 정의처럼 주가의 상승압력과 하락압력간의 상대적인 강도를 쉽게 보여준다. RSI가 클수록 주가의 상승추세가 크다는 뜻이며, 반대로 작을 수록 주가의 하락추세가 크다는 뜻이다. RSI 공식은 수학적으로 0과 1사이의 값을 만들게 한다. 만약 평균 주가상승폭(AU)이 평균 주가하락폭(AD)보다 훨씬 크다면 RSI 값은 1로 수렴할 것이며, 반대로 AU가 AD보다 훨씬 작다면 RSI값은 0으로 수렴할 것이다. 
이 값을 퍼센트로 환산하면, RSI 값은 0%~100% 사이의 값에서 움직이는 것이다.

그렇다면 N은 어떻게 구할까?

N은 사용자가 임의로 정할 수 있는 파라미터지만, RSI를 개발한 윌레스 와일더는 N을 **14일로 사용**할 것을 권장한다. RSI 역시 MACD와 같이 시그널(Signal)선이 별도로 존재한다. RSI 자체의 이동평균선이다. 말그대로 이동평균선이기 때문에 RSI보다 반응이 느리다는 특징이 있다. 기본적으로는 RSI의 6일 이동평균선을 사용한다.

· RSI 지표 읽는 법

 

① RSI는 50%를 기준으로, 50% 이상은 매수세 우세, 50% 이하는 매도세 우세를 나타냄.

② RSI가 70% 이상이면 매수세가 상당히 커서 초과매수 국면에 돌입.

③ RSI가 30% 이하이면 매도세가 상당히 커서 초과매도 국면에 돌입.

④ RSI가 시그널선을 상향돌파하면 단기적으로 매수세가 늘어나는 추세.

⑤ RSI가 시그널선을 하향돌파하면 단기적으로 매도세가 늘어나는 추세.

```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import datetime
 
ticker = ['011930']
df = fdr.DataReader('011930', '2021-01-01')

 
def calcRSI(df, period):
    U = np.where(df.diff(1)['Close'] > 0, df.diff(1)['Close'], 0) #df.diff를 통해 (기준일 종가 - 기준일 전일 종가)를 계산하여 0보다 크면 증가분을 감소했으면 0을 넣어줌
    D = np.where(df.diff(1)['Close'] < 0, df.diff(1)['Close'] *(-1), 0) #df.diff를 통해 (기준일 종가 - 기준일 전일 종가)를 계산하여 0보다 작으면 감소분을 증가했으면 0을 넣어줌
    AU = pd.DataFrame(U, index=df.index).rolling(window=period).mean() #AU, period=14일 동안의 U의 평균
    AD = pd.DataFrame(D, index=df.index).rolling(window=period).mean() #AD, period=14일 동안의 D의 평균
    RSI = AU / (AD+AU) * 100 #0부터 1로 표현되는 RSI에 100을 곱함
    return RSI
 
df.insert(len(df.columns), "RSI", calcRSI(df,14)) #web.DataReader를 통해 받았던 원래 DataFrame에 'RSI'열을 추가
df.insert(len(df.columns), "RSI signal", df['RSI'].rolling(window=9).mean()) #RSI signal(RSI 이동평균)을 구해서 추가함
print(df) #print로 잘 들어갔는지 출력
```

##2-2 Stochastic 

###상대적 강도 지수(Relative Strength Index)와 스토캐스틱 지표의 차이점
RSI는 가격 움직임의 속도를 측정하기 위해 설계되었지만, Stochastic은 시장이 일관된 범위에서 거래될 때 가장 잘 작동한다. 일반적으로 말하면, RSI는 동향 시장에서 더 유용하며 stochastic은 횡방향 또는 불안정한 시장에서 더 유용하다.

죠지 레인에 의해서 개발된 스톡캐스틱 기법은 가격이 하락기에 있는 경우 종가는 그날 형성된 변동폭 중 저가에 근접하는 경향이 강하고, 상승기에 있는 경우 종가는 그날 형성된 변동폭 중 고가에 근접하는 경향이 크다.
종가가 그날의 변동폭 중에서 저가에 근접하게 되면 하강 추세에 진입하게 되고 이런 현상이 몇일 계속되어 매도 분위기가 과열(OVER-SOLD)되면 추세의 전환이 이루어지고 다시 종가가 그날의 변동폭 중에서 고가에 접근하게 되어 상승추세에 진입하게 된다. 이런 분위기가 몇일 계속되면 매수 분위기가 과열 (OVER-BOUGHT) 되고 다시 상승 추세로의 전환이 이루어지게 된다.

이 기법은 %K와 %D의 두가지 인디케이터를 사용하게 된다. %K는 최근 5일간의 시장의 변동폭(고가-저가)중에서 오늘 종가의 위치를 나타내는 것이다. 만약 오늘의 고가가 지난 5일간의 저가와 일치하면 0 이 나오게 된다.(%K)

스토캐스틱 지표는 현재 종가에서 해당 기간의 최저값을 빼서 해당 기간의 최고값과 최저값의 차이로 나눈 다음 100을 곱하여 계산된다다. 예컨데, 14일 최고가가 15,000원, 최저값이 12,500원, 현재 종가는 14,500원이라면, 스토캐스틱 지표는 (14,500-12,500) / (15,000-12,500) * 100으로 이를 계산하면 80이 된다다.

현재 가격을 표준 기간의 범위와 비교함으로써 스토캐스틱 지표는 가격이 최근 고가 또는 저가 근처에서서 마감되는 일관성을 갖는다. 값이 80이면 자산이 과매수 직전에 있음을 의미한다.

```python
import pandas as pd
import matplotlib.pyplot as plt

ticker = ['011930']
df = fdr.DataReader('011930', '2021-01-01')

#일자(n, m, t)에 따른 Stochastic Fast와 Stochastic Slow를 계산하는 함수를 만들어 보자
def Stochastic(df, n=14, m=5, t=5):
    # 입력받은 값이 dataframe이라는 것을 정의해줌
    df = pd.DataFrame(df)

    # n일중 최고가
    ndays_high = df.High.rolling(window=n, min_periods=5).max()
    
    # n일중 최저가
    ndays_low = df.Low.rolling(window=n, min_periods=5).min()

    # Fast%K 계산
    fast_k = ((df.Close - ndays_low) / (ndays_high - ndays_low)) * 100

    # Fast%D (=Slow%K) 계산
    slow_k = fast_k.ewm(span=m).mean()

    # Slow%D 계산
    slow_d = slow_k.ewm(span=t).mean()

    # dataframe에 컬럼 추가
    df = df.assign(fast_k=fast_k, fast_d=slow_k, slow_k=slow_k, slow_d=slow_d)

    return df
```

```python
import yfinance as yf
import numpy as np
import ta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
import seaborn as sns

df = yf.download('META',start='2019-12-01')

# SMA 5days
sma5 = df['Close'].rolling(window=5).mean() #5일동안의 종가를 평균을 낸 것이 rolling(5).mean()
df_sma5 = sma5.to_frame(name='SMA5') # Series to DataFrame
# SMA 20days
sma20 = df['Close'].rolling(window=20).mean()
df_sma20 = sma20.to_frame(name='SMA20') # Series to DataFrame
#SMA 60days
sma60 = df['Close'].rolling(window=60).mean()
df_sma60 = sma60.to_frame(name='SMA60') #Series to DataFrame
#SMA 120days
sma120 = df['Close'].rolling(window=120).mean()
df_sma120 = sma120.to_frame(name='SMA120') #Series to DataFrame


# Concatenate SMA5 and SMA20
df_sma = pd.concat([df['Close'], df_sma5, df_sma20,df_sma60,df_sma120], axis=1)
# Remove NaN
df_sma = df_sma.dropna()

#정배열일때를 확인하기 (SMA5 > SMA20 > SMA60 > SMA120)
df_sma['SMA5 - SMA20'] = np.where((df_sma['SMA5'] - df_sma['SMA20'] > 0),1,0)
df_sma['SMA20 - SMA60'] = np.where((df_sma['SMA20'] - df_sma['SMA60'] > 0),1,0)
df_sma['SMA60 - SMA120'] = np.where((df_sma['SMA60'] - df_sma['SMA120'] > 0),1,0)



''' 만약 매수를 진행했으면 매도 시그널이 있을때까지는 매수를 멈춘다(계속 0 대입). 
    만약 매수를 진행하지 않았으면 매수 시그널(buy=1)이 있을때까지 매도를 멈춘다(계속 0 대입) '''
buy = 0 #우선 buy에 0을 대입한다. buy가 0일때는 계속 매수 시그널을 찾는다
sell = 0  #우선 sell에 0을 대입한다. buy가 1이고 sell이 0일때는 계속 매도 시그널을 찾는다
buy_list = [] # 빈 리스트를 생성한다
sell_list = []
for i in range(len(df_sma)): #매수 시그널과 매도 시그널을 계산해주기 위해 for문을 돌린다!
  if buy == 0:
    #매수 시그널을 찾아야한다. 정배열을 이룰때가 매수 시그널이다.
    buy = (df_sma['SMA5 - SMA20'][i]) & (df_sma['SMA20 - SMA60'][i]) & (df_sma['SMA60 - SMA120'][i])
    buy_list.append(buy)
    sell_list.append(0)
  elif buy == 1:
    #만약 SMA5 가 SMA20을 데드크로스 하면 매도한다! (1이 매도 타이밍)
    if df_sma['SMA20 - SMA60'][i] == 0: 
      sell = 1
    sell_list.append(sell)
    buy_list.append(0)
    if sell == 1:
      buy = 0
      sell = 0


df_sma['Buy'] = np.array(buy_list) #0과 1의 조합인 list를 array 형태로 변환한다.
df_sma['Sell'] = np.array(sell_list)

Buying_dates, Selling_dates = [],[]

for i in range(len(df_sma) - 1):
  if df_sma.Buy.iloc[i]: # 만약 buy = 1이라면,
    '''buy = 1인 날의 다음 날에 매수를 한다. 당일에 사게 되면 
    예견 편향(Look-ahead Bias)이 발생하게 된다. 
    예견 편향은 그때 당시에 알 수 없었던 정보나 데이터를 가지고 백테스팅을 함으로써 발생하게 되는 오류이다.
    따라서 특정 날에 매수 신호가 발생하면 다음날 시가에 매수를 하기 위해 df_sma.iloc[i+1] 을 적용해준다.
    '''
    Buying_dates.append(df_sma.iloc[i+1].name)
    for num,j in enumerate(df_sma.Sell[i:]):
      if j:
        Selling_dates.append(df_sma.iloc[i + num + 1].name)
        break
     
    
frame = pd.DataFrame({'Buying_dates': Buying_dates, 'Selling_dates': Selling_dates})
```

```python
def profitcalc():
  Buyprices = df.loc[frame.Buying_dates].Open
  Sellprices = df.loc[frame.Selling_dates].Open
  return (Sellprices.values - Buyprices.values) / Buyprices.values

profits = profitcalc()
profits

import matplotlib.pyplot as plt
print('총 수익률: ' + str((round((profits+1).prod(),4) -1) * 100) + ' %')
plt.figure(figsize=(20,10))
plt.plot(df.Close,color='k',alpha=0.7)
plt.scatter(frame.Buying_dates,df.Open[frame.Buying_dates],marker='^',color='g',s=500)
plt.scatter(frame.Selling_dates,df.Open[frame.Selling_dates],marker='v',color='r',s=500)
```


#### Ex2) 중급 백테스팅 전략: RSI + MACD + STOCHASTIC 활용

코드는 Ex1) 과 거의 유사합니다!!!

```python
import yfinance as yf
import numpy as np
import ta

df = yf.download('AMZN',start='2022-12-01',interval='1h')

#Stochastic 지표 
'''참고로 df['High'] 과 df.High 는 똑같다'''
df['%K'] = ta.momentum.stoch(df.High,df.Low,df.Close,window=14,smooth_window=3)
df['%D'] = df['%K'].rolling(3).mean()

#RSI 지표 
df['rsi'] = ta.momentum.rsi(df.Close,window=14)

#MACD 지표
df['macd'] = ta.trend.macd_diff(df.Close)

df.dropna(inplace=True)
```

```python
#스토케스틱의 매수 신호가 나온 시점부터 +3시간은 lagging
dfx = pd.DataFrame()
for i in range(1,4):
  mask = (df['%K'].shift(i) < 20) & (df['%D'].shift(i)<20)
  dfx = dfx.append(mask,ignore_index = True)
```

this column is accounting for the fact if the cross bellow 20 has occured. 

two rows, no cross, three rows, no cross

vertical sum > 0 이면 Stochastic 지표에 따라서 buy signal이다!

```python
def gettriggers(df,lags,buy=True):
  dfx=pd.DataFrame()
  for i in range(1,lags+1):
    if buy:
      mask = (df['%K'].shift(i) < 20) & (df['%D'].shift(i)<20)
    else:
      mask = (df['%K'].shift(i) > 85) & (df['%D'].shift(i)> 85)
    dfx=dfx.append(mask,ignore_index=True)
  
  return dfx.sum(axis = 0) #vertical sum의 합을 리턴해줘야 buy signal인지 아닌지 알 수 있다.
```

```python
df['Buytrigger'] = np.where(gettriggers(df,4,),1,0)
```

```python
df['Selltrigger'] = np.where(gettriggers(df,4, False),1,0)
```

```python
#매수 시그널과 매도 시그널을 계산해주기 위해 for문을 돌린다!

''' 만약 매수를 진행했으면 매도 시그널이 있을때까지는 매수를 멈춘다(계속 0 대입). 
    만약 매수를 진행하지 않았으면 매수 시그널(buy=1)이 있을때까지 매도를 멈춘다(계속 0 대입) '''

buy = 0
buy_list = []
sell_list = []
for i in range(len(df)):
  if buy == 0:
    #매수 시그널을 찾아야한다. 매도 시그널은 계속 0을 넣어준다,
    buy = (df.Buytrigger[i]) & (df['%K'].between(20,80)[i]) & (df['%D'].between(20,80)[i]) & (df.rsi[i]>30) & (df.macd[i]> 0 )
    buy_list.append(buy)
    sell_list.append(0)
  elif buy == 1:
    #매도 시그널을 찾아야한다. 매수 시그널은 계속 0을 넣어준다,
    sell = (df.Selltrigger[i]) & (df['%K'].between(20,80)[i]) & (df['%D'].between(20,80)[i]) &(df.rsi[i]<60) & (df.macd[i]< 0 )
    sell_list.append(sell)
    buy_list.append(0)
    if sell == 1:
      buy = 0


df['Buy'] = np.array(buy_list)
df['Sell'] = np.array(sell_list)
```

```python
Buying_dates, Selling_dates = [],[]

for i in range(len(df) - 1):
  if df.Buy.iloc[i]: # if contains a buy signal
    # Buy at the next time, because otherwise I would have a forward looking bias
    Buying_dates.append(df.iloc[i+1].name)
    for num,j in enumerate(df.Sell[i:]):
      if j:
        Selling_dates.append(df.iloc[i + num + 1].name)
        break
     
    
```

```python
#불필요한 포지션 삭제하기 ( 매도 없이 매수만 진행한 포지션 )
cutit = len(Buying_dates) - len(Selling_dates)
```

```python
if cutit:
  Buying_dates = Buying_dates[:-cutit]
  print(Buying_dates)
frame = pd.DataFrame({'Buying_dates': Buying_dates, 'Selling_dates': Selling_dates})

frame
```

```python
def profitcalc():
  Buyprices = df.loc[frame.Buying_dates].Open
  Sellprices = df.loc[frame.Selling_dates].Open
  return (Sellprices.values - Buyprices.values) / Buyprices.values
```

```python
profits = profitcalc()
profits

import matplotlib.pyplot as plt
print('총 수익률: ' + str((round((profits+1).prod(),4) -1) * 100) + ' %')
plt.figure(figsize=(20,10))
plt.plot(df.Close,color='k',alpha=0.7)
plt.scatter(frame.Buying_dates,df.Open[frame.Buying_dates],marker='^',color='g',s=500)
plt.scatter(frame.Selling_dates,df.Open[frame.Selling_dates],marker='v',color='r',s=500)
```
