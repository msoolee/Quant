## High-Yield Bond Spread
하이일드 채권 스프레드란 미국의 고수익 회사채(투자부적격신용등급, 정크본드)와 미국국채 간의 수익률 차이를 의미한다.  

하이일드 채권 스프레드가 확대되었다는 것은 신용등급이 낮은 기업들이 자금을 조달하기 매우 어려운 상황이라는 것을 의미한다.  

즉, 만약 하이일드 금리는 상승하고, 국채금리는 하락한다면 이는 곧 미국의 신용리스크 확대를 의미하며, 실물경제에 악영향을 미치게 된다.  

역사적으로도 닷컴버블, 글로벌 금융위기, 코로나 위기 구간에서 높은 하이일드 스프레드를 확인할 수 있다.

image.png

왼쪽 음영부터 1. 닷컴버블, 2.글로벌 금융위기, 3. 코로나 위기 순이다.  

그러므로 이를 이용해서 경기 침체 구간에서는 실물자산 비중을 높이고,  
스프레드가 정상궤도에 올랐을 때 투자자산 비중을 높이는 전략을 취해본다.

```python
# 필요 모듈 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import FinanceDataReader as fdr
```

```python
# TICKER 별 데이터 불러오기 및 시각화

# 하이일드 스프레드와 미국장기채 / 미국주식시장간의 관계
df = pd.DataFrame()

# 데이터 프레임 구하기
tickers = ['TLT', 'SPY']

for ticker in tickers:
  df[ticker] = fdr.DataReader(ticker, '2010')['Close'].resample('1M').last()

spread = fdr.DataReader("FRED:BAMLH0A0HYM2",'2004').resample('1M').last() # 하이일드 채권스프레드
mean = spread['BAMLH0A0HYM2'].rolling(window=12).mean() 
std = spread['BAMLH0A0HYM2'].rolling(window=12).std()
spread['UP_BAND'] = mean + 1.28*std # 신뢰구간 90% 상방구간

df[['SPREAD', 'UP_BAND']] = spread[['BAMLH0A0HYM2', 'UP_BAND']]

# 월별 변화량 구하기
monthly_returns = df.copy()
monthly_returns = monthly_returns.pct_change().fillna(0)
monthly_returns.iloc[0] = 0

# 상관계수 출력
print(monthly_returns.corr(method="pearson"))

# 데이터 시각화
for ticker in tickers:
  df.plot(y=ticker, figsize=(10,6))
  ax = df['SPREAD'].plot(secondary_y=True)
  plt.title(ticker + '& High Yield Spread')
  
plt.show()
```

```python
# 하이일드 채권스프레드
spread['2010':].plot()
```

## 투자 메서드:
결과적으로, SPREAD가 주식 수익률과 강한 음의 상관관계, 채권 수익률과는 양의 상관성을 가지므로,  
Spread가 최근 3년 기준 90% 신뢰구간을 이탈할 시에 경기침체 우려구간 이라고 판단,  
주식비중을 극단적으로 줄이고 국채나 금 등의 안전자산의 비중을 높인다.  

```python
# UP_BAND 돌파 스프레드를 0으로 표시

dates = df[df['SPREAD'] >= df['UP_BAND']].index
df.loc[dates, 'SPREAD'] = 0
```

**내 투자 종목:**


경기민감

- XLK 기술주

- XRT 소매

- XLB 원자재

- XLY 소비업종


크게 민감 X

- GLD 금

- XLE 에너지

- TLT 채권

```python
# 티커 리스트 정의
tickers = ['XLK', 'XRT', 'XLB', 'XLY', 'TLT', 'GLD', 'XLE']

# 데이터프레임 추가
data = pd.DataFrame()
for i in tickers:
  data[i] = fdr.DataReader(i,'2010-01-01')['Close'].resample('M').last()
 
data = data.fillna(0).resample('M').last()

# 수익률정의 및 전처리
rets = data.pct_change().fillna(0)
rets.replace([np.inf, -np.inf], 0, inplace=True)

# 연환산 기대수익률
er = rets.mean() * 12

# 수익률 공분산
cov = rets.cov() * 12
```

## 포트폴리오 구성

### 동일비중 포트폴리오

먼저, 자산비중을 정하기 전에 벤치마크로 이용할 동일비중 포트폴리오를 구현해본다. 
 

```python
# Non-Rebalancing equally weights
noa = len(tickers)
e_weights = np.array([1/noa for _ in range(noa)])
risk = np.sqrt(e_weights.T @ cov @ e_weights)

ew_port_cum_rets = (rets @ e_weights +1).cumprod() -1
print('총 수익률: ' + str(round((ew_port_cum_rets[-1]) * 100 , 3)) + '% / 변동성(연간):' + str(round(risk, 3)))
plt.plot(ew_port_cum_rets) # 시각화
plt.title('EW Portfolio')
plt.xlabel('Date')
plt.ylabel('Returns')
```

동일비중 포트폴리오에 하이일드 스프레드 마켓타이밍 전략을 섞어본다.  

동일비중으로 투자하다가, 하이일드가 정상수준을 벗어날 시  
주식 비중을 줄이고 채권과 금의 투자비중을 늘림림

```python
# Rebalancing equally weights
# 주식 비중 30% 감소, 금 채권비중을 그만큼 추가
normal_weights = [0.15, 0.15, 0.14, 0.14, 0.14, 0.14, 0.14]
abnormal_weights = [0.1, 0.1, 0.1, 0.1, 0.33, 0.33, 0.14]
weight_list = []
for i in df['SPREAD']:
  # i == 0 -> SPREAD가 최근 1년 내 신뢰구간 90% 상향이탈
  if (i == 0.0):
    weight_list.append(abnormal_weights) 
  else:
    weight_list.append(normal_weights)

total_weight = np.array(weight_list)
w_df = pd.DataFrame(total_weight, index=rets.index, columns=rets.columns)
returns = rets * w_df.shift(1)
rew_port_rets = returns.sum(axis=1)
rew_port_cum_rets = (1 + rew_port_rets).cumprod() - 1

risk = rew_port_rets.std() * np.sqrt(12)

print('총 수익률: ' + str(round((rew_port_cum_rets[-1]) * 100 , 3)) + '% / 변동성(연간):' + str(round(risk, 3)))

plt.plot(rew_port_cum_rets) # 시각화
plt.title('Reb-EW Portfolio')
plt.xlabel('Date')
plt.ylabel('Returns')
```

### 90/10

워렌버핏이 주장한 90/10 포트폴리오를 조금 변형해서 사용한다.  
원래의 의미는 은퇴자금을 투자하려는 투자자들에게 주식 인덱스에 90%를, 채권에 10%를 투자하고 잊어라!  
라는 의미였지만, 마켓타이밍 전략을 이용해 위험을 조금은 헷지할 수 있지 않을까 싶어  
주식 인덱스 ETF를 섹터별 ETF와 같이 다각화 시키고, 10%의 비중을 채권, 금, 에너지 ETF에 투자하도록 한다.  


먼저, 마켓타이밍 전략이 제대로 통하나 싶어 먼저 리벨런싱 없이 모든 기간에 가중치를 처음 주어진 그대로 가져간다.  


```python
# Non-Rebalancing 90/10 weights
weights = np.array([0.225, 0.225, 0.225, 0.225, 0.05, 0.025, 0.025])
wb_port_cum_rets = (rets @ weights +1).cumprod() -1

risk = np.sqrt(weights.T @ cov @ weights)

print('총 수익률: ' + str(round((wb_port_cum_rets[-1]) * 100 , 3)) + '% / 변동성(연간):' + str(round(risk, 3)))
plt.plot(wb_port_cum_rets) # 시각화
plt.title('WB Portfolio')
plt.xlabel('Date')
plt.ylabel('Returns')
```

동일 비중 포트폴리오에 비해 상당히 개선된 수익률을 보여준다.  
주식의 비중을 크게 높인 만큼 변동성도 커진 것으로 확인된다.  

다음은 하이일드 스프레드를 통해 비중을 달리하는 포트폴리오를 구현해본다.  
경기침체 우려가 예상되는 시기에는 주식 비중을 0으로 하고, 채권과 금, 에너지에 각각 가중치를 부여한다.  
신뢰구간 90%를 벗어났다는 것이 이렇게까지 극단적으로 대처하는 이유이다.

```python
# Rebalancing 90/10 비중
abnormal_weights = [0.0, 0.0, 0.0, 0.0, 0.4, 0.5, 0.1]
normal_weights = [0.225, 0.225, 0.225, 0.225, 0.05, 0.025, 0.025]

weight_list = []

for i in df['SPREAD']:
  # i == 0 -> SPREAD가 최근 1년 내 신뢰구간 90% 상향이탈
  if (i == 0.0):
    weight_list.append(abnormal_weights) 
  else:
    weight_list.append(normal_weights)

total_weight = np.array(weight_list) #2010년 1월 가중치부터 현재 가중치를 total_weight array에 포함시킨다
```

```python
w_df = pd.DataFrame(total_weight, index=rets.index, columns=rets.columns) #total_weight array를 DataFrame으로 바꿔줘서 rets 값과 곱하기가 가능하도록 만들어주기기
returns = rets * w_df.shift(1)
rwb_port_rets = returns.sum(axis=1)
rwb_port_cum_rets = (1 + rwb_port_rets).cumprod() - 1
risk = rwb_port_rets.std() * np.sqrt(12)
```

```python
print('총 수익률: ' + str(round((rwb_port_cum_rets[-1]) * 100 , 3)) + '% / 변동성(연간):' + str(round(risk, 3)))
plt.plot(rwb_port_cum_rets) # 시각화
plt.title('Reb-WB portfolio')
plt.xlabel('Date')
plt.ylabel('Returns')
```

최종적으로 마켓타이밍 전략을 사용하기 전보다 나은 수익률을 보여준다.  


마지막으로 네 그래프들을 각각 비교해보며 마무리한다.  


```python
def get_improve(bench, port):
  improve = (port[-1] / bench[-1] - 1) * 100
  improve = improve.round(2)
  return improve


fig, ax = plt.subplots(3,2, figsize=(16,14))

ax[0,0].plot(ew_port_cum_rets.index, ew_port_cum_rets, '--b', label='Benchmark')
ax[0,1].plot(ew_port_cum_rets.index, ew_port_cum_rets, '--b', label='Benchmark')
ax[1,0].plot(ew_port_cum_rets.index, ew_port_cum_rets, '--b', label='Benchmark')
ax[1,1].plot(rew_port_cum_rets.index, rew_port_cum_rets, '--b', label='Benchmark')
ax[2,0].plot(rew_port_cum_rets.index, rew_port_cum_rets, '--b', label='Benchmark')

ax[0,0].set_title(f'EW - REW / improved: {get_improve(ew_port_cum_rets, rew_port_cum_rets)} %')
ax[0,1].set_title(f'EW - WB / improved: {get_improve(ew_port_cum_rets, wb_port_cum_rets)} %')
ax[1,0].set_title(f'EW - RWB / improved: {get_improve(ew_port_cum_rets, rwb_port_cum_rets)} %')
ax[1,1].set_title(f'REW - WB / improved: {get_improve(rew_port_cum_rets, wb_port_cum_rets)} %')
ax[2,0].set_title(f'REW - RWB / improved: {get_improve(rew_port_cum_rets, rwb_port_cum_rets)} %')

ax[0,0].plot(rew_port_cum_rets.index, rew_port_cum_rets, '-r', label='My Port')
ax[0,1].plot(wb_port_cum_rets.index, wb_port_cum_rets, '-r', label='My Port')
ax[1,0].plot(rwb_port_cum_rets.index, rwb_port_cum_rets, '-r', label='My Port')
ax[1,1].plot(wb_port_cum_rets.index, wb_port_cum_rets, '-r', label='My Port')
ax[2,0].plot(rwb_port_cum_rets.index, rwb_port_cum_rets, '-r', label='My Port')
```
