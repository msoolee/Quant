## 1. 듀얼 모멘텀 전략 변형판
(주식)  
SPY = S&P500 지수  
XLK = 미국 IT 섹터  
XLV = 미국 헬스케어 섹터  
(채권)  
LQD = 미국 회사채  
TLT = 미국 장기채 (20years+)  
HYG = 미국 고수익채권  
(현금)  
BIL = 미국 초단기채 (1~3개월 채권) = 현금   

### (1) 전략 구성 논리  
지난주에 발표했던 듀얼 모멘텀 전략을 약간 변형시켜봤습니다  

__방법__  
(1) 각 자산별로 아래의 공식에 따라 모멘텀을 구합니다  
(2) 3개월마다 한번씩 찾아오는 월말 리밸런싱일에 SPY의 모멘텀이 BIL의 모멘텀보다 크거나 같다면, 상승장으로 판단하고 가장 높은 모멘텀을 가진 주식 1개에 투자합니다  
(3) 반대로 SPY의 모멘텀이 BIL의 모멘텀보다 작다면, 하락장으로 판단하고 채권과 현금에 투자합니다 투자하는 방법은 최대 3개의 자산을 선택하며, BIL의 모멘텀보다 높은 모멘텀을 보이는 채권에만 투자하고 나머지는 모두 현금 BIL에 투자합니다  
그에 따라 현금 비중은 0% = 채권 3개, 33% = 채권 2개, 현금 1개, 66% = 채권 1개, 현금 2개, 100% = 현금 3개 중 하나에 해당합니다
  
__모멘텀__ = 각 자산의 6개월, 12개월 모멘텀의 평균값  

__N개월 모멘텀 계산식__  
(오늘 종가 / N개월 전 종가) - 1 = N개월 수익률  

__횡적 배분모형__: 듀얼 모멘텀 변형  
(원래는 '샤프지수 최적화 + 듀얼 모멘텀 변형'을 쓰려고 했는데, 샤프비율 최적화하니 SPY로 가중치가 몰리는 코너해 문제가 발생해서 사용하지 않았습니다)

__종적 배분모형__: 위의 방법 (3)처럼 하락장일 경우, BIL의 모멘텀보다 높은 모멘텀을 보이는 채권에만 투자하고 나머지는 모두 현금에 투자합니다  

__마켓 타이밍__: 듀얼 모멘텀 변형

### (2) 자산 선택
주식은 듀얼 모멘텀 전략 자산 중 하나인 SPY, 그리고 SPY와 상관계수가 높고 TLT와는 상관계수가 낮으면서도 고수익을 보이는 XLK와 XLV에 투자했습니다 (heatmap 활용)  
채권은 수익과 변동성이 비교적 높은 LQD, TLT, HYG로 골랐습니다  
현금은 포트폴리오의 변동성을 헤지하기 위해 BIL로 골랐습니다

```python
#패키지 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime, timedelta
from scipy.optimize import minimize
sns.set()
```

## 2. 동일비중 포트폴리오
우선 동일비중으로 투자했을 때, 수익률을 보겠습니다. (3개월 리밸런싱)

```python
ew_tickers = ['SPY', 'XLK', 'XLV', 'TLT', 'LQD', 'HYG', 'BIL']
ew_etf = yf.Tickers(ew_tickers)
ew_data = ew_etf.history(start='2010-01-01',actions= False)['Close']
ew_data = ew_data.resample('3BM').last()

# 수익률 계산
ew_rets = ew_data.pct_change().fillna(0)

# 동일비중
ew_port_rets = ew_rets.mean(axis =1)

# 포트폴리오 누적 수익률
ew_port_cum_rets = (1 + ew_port_rets).cumprod() - 1
ew_port_cum_rets = ew_port_cum_rets * 100
# 총 수익률
ew_port_cum_rets_end = str(round(ew_port_cum_rets[-1], 3))

# 그래프 결과
plt.plot(ew_port_cum_rets)
plt.title('EW Backtest')
plt.xlabel('Date')
plt.ylabel('Returns')
print('총 수익률: ' + ew_port_cum_rets_end + '%')
```

## 3. 변형한 듀얼 모멘텀 전략
오리지널 듀얼 모멘텀 전략을 변형한 전략을 백테스트해보겠습니다. (3개월 리밸런싱)

```python
tickers = ['SPY', 'XLK', 'XLV', 'TLT', 'LQD', 'HYG', 'BIL']
etf = yf.Tickers(tickers)

# 모멘텀 계산하기 위해 1년이 필요해서
# 2009년 데이터부터 다운로드
data = etf.history(start='2009-01-01', actions=False)
data.drop(['Open', 'High', 'Low', 'Volume'], inplace=True, axis=1)
data = data.droplevel(0, axis=1)

# 모멘텀 계산 함수
def get_momentum(x):
    temp_list = [0 for i in range(len(x.index))]
    momentum = pd.Series(temp_list, index=x.index)

    try:    
        # 6, 12개월 평균 모멘텀 사용
        # N개월 모멘텀 = (오늘 종가 / N개월 전 종가) - 1 = N개월 수익률
        before6 = data[x.name-timedelta(days=185):x.name-timedelta(days=180)].iloc[-1]
        before12 = data[x.name-timedelta(days=370):x.name-timedelta(days=365)].iloc[-1]
        
        # 평균 모멘텀 = (6개월 모멘텀 + 12개월 모멘텀) / 2
        momentum = ((x / before12 - 1) + (x / before6 - 1)) / 2

    
    # 2009-01~2009-07 6개월치 데이터가 안 쌓여 있어서
    # 예외, 오류 발생 -> pass로 무시
    except:
        pass
    
    # 계산한 momentum 값 반환
    return momentum

# 주식 모멘텀 지수(M)
# 원래 데이터프레임 'data' 에 (자산 티커_M)으로 자산 모멘텀 열 추가
mom_col_list = [col+'_M' for col in data.columns]
data[mom_col_list] = data.apply(lambda x: get_momentum(x), axis=1)

# 백테스팅 기간 설정
# 2009년부터 모멘텀 계산 시작 -> 2010년부터 투자
start_day = datetime(2010,1,1)
end_day = datetime(2023,1,31)
data = data[start_day:end_day]

# 분기말 데이터 추출
# rule = '3BM' : 3 Business Month, 휴일 제외 영업일 3개월별 데이터
data = data.resample(rule='3BM').last()

# 결측값 -> 0
data = data.fillna(0)
```

```python
# 듀얼 모멘텀 전략에 따라 자산 선택
def select_asset(x):
    
    # 자산, 가격 데이터 저장하기 위해 asset 시리즈 선언
    asset = pd.Series([0,0,0,0,0,0], index=['ASSET1','PRICE1','ASSET2','PRICE2','ASSET3','PRICE3'])
    
    # 3개 자산 선택
    # 원래는 SPY의 모멘텀이 0 보다 클 때는 주식 1개, 0 보다 작아서 채권에 투자할 때 채권 3개 선택이지만
    # 편의상 3개 자산 선택으로 통일함
    # 1개 선택하는 경우에는 3개 모두 같은 자산으로 선택
    momentum1 = None
    momentum2 = None
    momentum3 = None
    
    # 상위 채권 고르기 위해 모멘텀 내림차순 정렬
    sorted_momentum = x[['TLT_M', 'LQD_M', 'HYG_M', 'BIL_M']].sort_values(ascending=False)
    
    # 상위 주식 고르기 위해 모멘텀 내림차순 정렬
    sorted_momentum1 = x[['SPY_M', 'XLK_M', 'XLV_M']].sort_values(ascending=False)
    
    # SPY 수익률 >= BIL 수익률
    # 상승장 = 모멘텀 상위 주식 1개 투자
    if x['SPY_M'] >= x['BIL_M']:
        momentum1 = sorted_momentum1[0]
        momentum2 = sorted_momentum1[0]
        momentum3 = sorted_momentum1[0]
        
        # 선택한 자산의 이름과 가격 저장
        asset['ASSET1'] = x[x == momentum1].index[0].split('_')[0]
        asset['PRICE1'] = x[asset['ASSET1']]
        asset['ASSET2'] = x[x == momentum2].index[0].split('_')[0]
        asset['PRICE2'] = x[asset['ASSET2']]
        asset['ASSET3'] = x[x == momentum3].index[0].split('_')[0]
        asset['PRICE3'] = x[asset['ASSET3']]
    
    # SPY 수익률 < BIL 수익률
    # 하락장 = 채권과 현금 투자
    else:
        # BIL 수익률 > 상위 3개 채권 수익률
        # 모두 현금 보유(BIL)
        if x['BIL_M'] == sorted_momentum[0]:
            momentum1 = sorted_momentum[0]
            momentum2 = sorted_momentum[0]
            momentum3 = sorted_momentum[0]
            
            asset['ASSET1'] = x[x == momentum1].index[0].split('_')[0]
            asset['PRICE1'] = x[asset['ASSET1']]
            asset['ASSET2'] = x[x == momentum2].index[0].split('_')[0]
            asset['PRICE2'] = x[asset['ASSET2']]
            asset['ASSET3'] = x[x == momentum3].index[0].split('_')[0]
            asset['PRICE3'] = x[asset['ASSET3']]
        
        # 상위 3개 중 2개가 BIL 수익률보다 낮을 때
        # 상위 채권 1개 투자, 나머지 두개는 현금 보유
        # 채권 33%
        elif x['BIL_M'] == sorted_momentum[1]:
            momentum1 = sorted_momentum[0]
            momentum2 = sorted_momentum[1]
            momentum3 = sorted_momentum[1]
            
            asset['ASSET1'] = x[x == momentum1].index[0].split('_')[0]
            asset['PRICE1'] = x[asset['ASSET1']]
            asset['ASSET2'] = x[x == momentum2].index[0].split('_')[0]
            asset['PRICE2'] = x[asset['ASSET2']]
            asset['ASSET3'] = x[x == momentum3].index[0].split('_')[0]
            asset['PRICE3'] = x[asset['ASSET3']]
        
        # 상위 3개 중 1개가 BIL 수익률보다 낮을 때
        # 상위 채권 2개 투자, 나머지 하나는 현금 보유
        # 채권 66%
        elif x['BIL_M'] == sorted_momentum[2]:
            momentum1 = sorted_momentum[0]
            momentum2 = sorted_momentum[1]
            momentum3 = sorted_momentum[2]
            
            asset['ASSET1'] = x[x == momentum1].index[0].split('_')[0]
            asset['PRICE1'] = x[asset['ASSET1']]
            asset['ASSET2'] = x[x == momentum2].index[0].split('_')[0]
            asset['PRICE2'] = x[asset['ASSET2']]
            asset['ASSET3'] = x[x == momentum3].index[0].split('_')[0]
            asset['PRICE3'] = x[asset['ASSET3']]
        
        # 상위 3개 채권의 모멘텀이 모두 0 보다 클 때
        # 상위 채권 3개 모두 투자
        # 채권 100%
        else:
            momentum1 = sorted_momentum[0]
            momentum2 = sorted_momentum[1]
            momentum3 = sorted_momentum[2]
            
            asset['ASSET1'] = x[x == momentum1].index[0].split('_')[0]
            asset['PRICE1'] = x[asset['ASSET1']]
            asset['ASSET2'] = x[x == momentum2].index[0].split('_')[0]
            asset['PRICE2'] = x[asset['ASSET2']]
            asset['ASSET3'] = x[x == momentum3].index[0].split('_')[0]
            asset['PRICE3'] = x[asset['ASSET3']]
            
    # asset 반환
    return asset

# apply 써서 함수 바로 적용
data[['ASSET1','PRICE1','ASSET2','PRICE2','ASSET3','PRICE3']] = data.apply(lambda x: select_asset(x), axis=1)

# IU = Investment Universe
# 각 자산별 수익률(P)
# data.columns[:7] = tickers 열까지 col 불러오기
IU = data.columns[:7]
return_col = [col + '_P' for col in data[IU].columns]
data[return_col] = data[IU].pct_change()

# 결측값 -> 0
data[return_col] = data[return_col].fillna(0)
```

```python
# 각 시점의 수익 RETURN
# 누적수익률 RETURN_ACC
data['RETURN'] = 0
data['RETURN_ACC'] = 0

for i in range(len(data)):
    strat_return = 0
    
    if i > 0:
        # 투자한 3개 자산의 평균 수익률
        strat_return = (data[data.iloc[i-1]['ASSET1'] + '_P'].iloc[i] + data[data.iloc[i-1]['ASSET2'] + '_P'].iloc[i] + data[data.iloc[i-1]['ASSET3'] + '_P'].iloc[i]) / 3
    data.loc[data.index[i], 'RETURN'] = strat_return    
    data.loc[data.index[i], 'RETURN_ACC'] = (1+data.loc[data.index[i-1], 'RETURN_ACC'])*(1+strat_return)-1

# 수익률에 100을 곱해서 백분율로 표기   
data[['RETURN','RETURN_ACC']] = data[['RETURN','RETURN_ACC']] * 100
data[return_col] = data[return_col] * 100

# 손실 구하기
data['BALANCE'] = (1+data['RETURN']/100).cumprod()
data['DRAWDOWN'] = -(data['BALANCE'].cummax() - data['BALANCE']) / data['BALANCE'].cummax()
data[['BALANCE','DRAWDOWN']] = data[['BALANCE','DRAWDOWN']] * 100
```

## 결과

```python
# 그래프로 그리기
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(data['RETURN_ACC'].index, data['RETURN_ACC'], '-b', label='Dual Momentum')
ax.plot(ew_port_cum_rets.index, ew_port_cum_rets, '--r', label='Benchmark')
leg = ax.legend();
plt.title('Portfolio Backtest')
plt.xlabel('Date')
plt.ylabel('Returns')

# 총 수익률
port_cum_rets_end = str(round(data['RETURN_ACC'][-1], 3))
print('총 수익률: ' + port_cum_rets_end + '%')

# 샤프비율
# 3개월 단위이므로 평균수익률에 4를 곱하고, 표준편차에 루트4를 곱하여 연율화함
sharpe_ratio = (data['RETURN'].mean() * 4) / (data['RETURN'].std() * np.sqrt(4))
print('샤프비율: {}'.format(round(sharpe_ratio, 2)))

# 변동성 = 연 표준편차
std = round(data['RETURN'].std() * np.sqrt(4), 2)
print('표준편차: {}%'.format(std))

# MDD = 최대손실
mdd = round(data['DRAWDOWN'].min(), 2)
print('MDD: {}%'.format(mdd))
```

총 수익률 187.2% -> 544.2%으로 약 357%p 개선하는 효과를 보였다.

## 상관계수 히트맵

```python
tickers = ['SPY', 'VEA', 'VWO', 'XLK', 'XLV', 'LQD', 'TLT', 'BIL', 'HYG']
etf = yf.Tickers(tickers)
data = etf.history(start='2010-01-01',actions= False)['Close']

rets = data.pct_change().fillna(0)

corr = rets.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(np.round(corr, 2), annot=True)
```

__XLK와 XLV를 주식 유니버스에 넣은 이유__  
오리지널 듀얼 모멘텀 전략에서 사용하는 SPY와 EFA(VEA와 동일)의 상관계수는 0.88이다.  
이와 비슷한 상관계수를 보이고 고수익에 해당하는 주식에 투자하고 싶었고, XLK와 XLV가 각각 0.93, 0.86으로 EFA와 비슷한 상관계수를 보이고, IT와 헬스케어 섹터로 고수익에 해당해 주식 유니버스에 추가했다.
